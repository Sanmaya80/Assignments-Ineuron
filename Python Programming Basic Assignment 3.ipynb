```python
# 1.Write a Python Program to Check if a Number is Positive, Negative or Zero?

num=int(input("Enter a Number:"))
if num>0:
    print('The Number {} is Positive Number'.format(num))
elif num<0:
    print('The Number {} is Negative Number',format(num))
elif num==0:
    print('The Number {} is Zero',format(num))

```

```python
#2.	Write a Python Program to Check if a Number is Odd or Even?

num=int(input("Enter the Number: "))
if num % 2==0:
    print("The Number {} is prime".format(num))
else:
    print("The Number {} is prime".format(num))
    
```

```python
# 3.Write a Python Program to Check Leap Year?

year=int(input("Enter the year :"))

if year % 4==0:
    if year % 100==0:
        if year % 400==0:
            print('The {} is a Leap year'.format(year))
        else:
            print('The {} is not a Leap year'.format(year))
    else:
        print('The {} is a Leap year'.format(year))
else:
    print('The {} is not a Leap year'.format(year))
```

```python
# 4.Write a Python Program to Check Prime Number?

num=int(input("Enter a Number : "))
if num>1:
    for i in range(2,num):
        if num % i==0:
            print('The {} is a not prime Number'.format(num))
            break
        else:
            print('The {} is a prime Number'.format(num))

    
```

```python
# 5.Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

for num in range(1,10000):
        if num >1:
            for i in range(2,num):
                if (num % i)==0:
                    break
                    
            else:
                print('The {} is a prime Number'.format(num))
```
