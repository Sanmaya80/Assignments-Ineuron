```python
#1.Write a Python Program to Find the Factorial of a Number?

num=int(input("Enter a Number :"))
fact=1
if num <0:
    print('Factorial of {} number does not exist'.format(num))
elif num==0:
    print('Factorial of {} is 1'.format(num))
else:
    for i in range(1,num+1):
        fact=fact*i
    print('Factorial os {} is {}'.format(num,fact))    
        
```

```python
# 4.Write a Python Program to Check Armstrong Number?

num=int(input("Enter a Number to check if it is a Armstrong Number :"))
sum=0
val=num
while val > 0:
        dig=val % 10
        sum +=dig ** 3
        val=val //10

if num==sum:
    print('The Number {} is a Armstrong Number'.format(num))
else:
    print('The Number {} is not a Armstrong Number'.format(num))
```

```python
# 5.Write a Python Program to Find Armstrong Number in an Interval?
lowint=int(input("Enter a lower interval :"))
upint=int(input("Enter a upper interval :"))

for num in range(lowint,upint+1):
    ordn=len(str(num))
    
    sum=0
    temp=num
    while temp>0:
        dig=temp % 10
        sum=sum+ dig**ordn
        temp=temp//10
        
    if num==sum:
        print(num)
```

```python
# 6.Write a Python Program to Find the Sum of Natural Numbers?

num=int(input("Enter a number :"))

sum=0

while num>0:
    sum=sum+num
    num=num-1
print('the sum is {}'.format(sum))
```
