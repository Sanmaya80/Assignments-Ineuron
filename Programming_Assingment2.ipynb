```python
#1.	Write a Python program to convert kilometers to miles?

kilometers = float(input("Enter value in kilometers: "))

factor_miles = 0.621371

miles = kilometers * factor_miles

print(kilometers ,' kilometers is equal to', miles, ' miles')
```

```python
#2.	Write a Python program to convert Celsius to Fahrenheit?

celsius=float(input("Enter value in celcius:"))

fahrenheit=(celsius*1.8)+32

print(celsius,' celsius is equal to ', fahrenheit, 'fahrenheit')
```

```python
#3.	Write a Python program to display calendar?

import calendar

year=int(input("Enter the year of Calendar:"))
month=int(input("Enter the year of Calendar:"))

print(calendar.month(year,month))
```

```python
#5.	Write a Python program to swap two variables without temp variable?
first_no=int(input("Enter the first number:"))
second_no=int(input("Enter the first number:"))
print("Before swapping", first_no,second_no)
first_no,second_no=second_no,first_no
print("After swapping",first_no,second_no)

```
